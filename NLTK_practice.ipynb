Importing Natural Language Toolkit

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import  word_tokenize
from nltk.tokenize.treebank import TreebankWordDetokenizer
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

Word Pre-processing:
Stemming

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.tokenize import word_tokenize
```

```python
#creating instance
ps = PorterStemmer()
ls = LancasterStemmer()
```

Now we will loop over the reviews and Stemm the words in it.

```python
for line in reviews:
  line = line.lower()
  token = word_tokenize(line)
  for w in token:
    print('Porter Stemmer-->'+ w, ':', ps.stem(w) )
    print('Lancaster Stemmer-->' + w ,':', ls.stem(w))
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer
#creating an instance
wn_lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
for lines in reviews:
  lines = lines.lower()
  token = word_tokenize(lines)
  for w in token:
    print('Lemmatizer-->' + w , ':', wn_lemmatizer.lemmatize(w)  )
```

Removing Stopwords

```python
english_stopwords = set(stopwords.words('english'))
```

```python
cleaned_corpus = []
for line in reviews:
  line = line.lower()
  token = word_tokenize(line)  #tokenizing the lines i.e converting them to words
  
  # creating a variable that will tokens that are free from stopwords
  token_without_stopwords = [word for word in token if not word in english_stopwords]
  
  #detokenizing the string and storing them in a variable called cleaned string
  cleaned_string = TreebankWordDetokenizer().tokenize(token_without_stopwords)

  #appending the cleaned detokenized string in the cleaned corpus
  cleaned_corpus.append(cleaned_string)
  cleaned_corpus 
```
